```python
import pandas as pd
import statsmodels.api as sm
from matplotlib import pyplot as plt  

df = pd.read_excel('2阳光信息.xlsx',usecols=['Y','红','气','车','白','橄','丑'])
# df = pd.read_excel('2阳光信息.xlsx',usecols=['Y','红','气','车','白','丑'])

X = df.iloc[:,1:]
y = -df['Y']

print(X)
print(y)
```

```python
X_sm = sm.add_constant(X)
model_sm = sm.OLS(y, X_sm).fit()
print(model_sm.summary())
```
