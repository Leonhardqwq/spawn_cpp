```python
import numpy as np
import pandas as pd
```

```python
cols_power = [18, 5, 8, 11, 17, 6,  1, 20]
df = pd.read_excel('0DOE.xlsx',sheet_name='DOE').iloc[1:,1:9]
```

```python
df
```

```python
df.replace(-1, 0, inplace=True)
```

```python
df['ans'] = df.apply(
    lambda row: sum([
        row[i] * (1 << (cols_power[idx])) 
        for idx, i in enumerate(df.columns)
    ]),
    axis=1
)
df['ans'] = df['ans'].apply(lambda x: f'0b{x:021b}')
```

```python
df
```

```python
for v in df['ans'].values:
    print(v, end=',\n')
```
