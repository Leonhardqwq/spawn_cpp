```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('DOE.xlsx')
df = df.drop(columns=df.columns[0])  # 删除第一列
given_design = df.to_numpy()

factors = len(given_design[0])
n_experiments = len(given_design)
n_candidates = 2 ** factors  

# 生成所有可能的实验点（候选集）
levels = np.array([-1, 1])
candidate_set = np.array(np.meshgrid(*[levels]*factors, indexing='ij')).T.reshape(-1, factors)
```

```python

def find_design_indices(design, candidate_set):
    """
    查找给定design在候选集中的索引
    
    参数:
    design (np.ndarray): 给定的实验设计矩阵，每行代表一个实验点
    candidate_set (np.ndarray): 候选集矩阵，包含所有可能的实验点
    
    返回:
    np.ndarray: 对应的索引数组
    """
    # 将候选集转换为元组列表，并构建索引映射
    candidate_tuples = [tuple(row) for row in candidate_set]
    index_map = {tuple(candidate): idx for idx, candidate in enumerate(candidate_set)}
    
    # 查找每个设计点在候选集中的索引
    indices = []
    for row in design:
        row_tuple = tuple(row)
        if row_tuple in index_map:
            indices.append(index_map[row_tuple])
        else:
            # 如果找不到该点，抛出异常或处理缺失值
            raise ValueError(f"设计点 {row} 不在候选集中")
    
    return np.array(indices)
```

```python
find_design_indices(given_design,candidate_set)
```
