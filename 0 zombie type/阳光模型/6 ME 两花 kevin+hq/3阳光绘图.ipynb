```python
import matplotlib.pyplot as plt

filename = 'output.txt'
with open(filename, 'r') as file:
    data = file.readlines()
data = [float(line.strip()) for line in data]

idx = 0
seeds = []
dic = {}
for i in range(20):
    seeds.append(data[idx])
    idx+=1
    cost_list = data[idx:idx+10000]
    idx+=10000
    dic[seeds[-1]] = cost_list

for key,value in dic.items():
    # print(key)    
    for i in range(10000):
        if value[i]<-500:
            tmp = max(0,i-50)
            dic[key] = value[tmp:tmp+100]
                        
            break
    
for key,value in dic.items():
    print(key,len(value))    
    plt.plot(value)
    plt.title(f'{key}')
    plt.xlabel('Level')
    plt.ylabel('Sun')
    plt.grid(True)

    plt.show()

```
