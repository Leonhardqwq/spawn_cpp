```python
import pandas as pd
import statsmodels.api as sm
from matplotlib import pyplot as plt  

df = pd.read_excel('2.0阳光信息.xlsx',usecols=['Y','红','车','橄','气','舞','杆','白','权重和'])
df['权重和']=df['权重和']/10000

df1 = df[(df['车'] == 1) & (df['橄'] == 1)]
df2 = df[~((df['车'] == 1) & (df['橄'] == 1))]
```

```python
def OLSmodel(df,const=True):
    X = df.iloc[:,1:]
    y = -df['Y']
    X_sm = sm.add_constant(X) if const else X
    model_sm = sm.OLS(y, X_sm).fit()
    print(model_sm.summary())
    return pd.DataFrame(model_sm.params)
```

```python
OLSmodel(df[['Y','车','橄','气','舞','杆','权重和']],False)
```

```python
OLSmodel(df1[['Y','气','白','权重和']],True)
```

```python
OLSmodel(df2[['Y','车','橄','气','舞','杆','白','权重和']],False)
```
