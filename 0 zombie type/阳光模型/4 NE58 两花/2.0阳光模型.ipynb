```python
import pandas as pd
import statsmodels.api as sm
from matplotlib import pyplot as plt  

# df = pd.read_excel('2.0阳光信息.xlsx',usecols=['Y','红','气','橄','舞','篮','撑','白','权重和'])
df = pd.read_excel('2.0阳光信息.xlsx',usecols=['Y','气','橄','舞','篮','撑','白','权重和'])
df['权重和']=df['权重和']/10000

X = df.iloc[:,1:]
y = -df['Y']

# print(X)
# print(y)
```

```python
X_sm = X
# X_sm = sm.add_constant(X)
model_sm = sm.OLS(y, X_sm).fit()
print(model_sm.summary())
```
