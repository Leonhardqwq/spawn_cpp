```python
import pandas as pd
import statsmodels.api as sm
from matplotlib import pyplot as plt  

df = pd.read_excel('2阳光信息.xlsx')
df['权重和']=df['权重和']/10000
df = df.drop('flag',axis=1)
df.columns
```

```python
df = df[[
    'Y', 
    '撑杆', 
    '橄榄', '舞王', 
    '红眼', 
    '权重和'
]]
```

```python

df = df[[
    'Y', 
    '橄榄', '舞王', 
    '白眼','红眼', 
    '权重和'
]]
```

```python
X = df.iloc[:,1:]
y = df['Y']

X,y
```

```python
X_sm = X
# X_sm = sm.add_constant(X)
model_sm = sm.OLS(y, X_sm).fit()

print(model_sm.summary())
```
