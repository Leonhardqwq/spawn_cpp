```python
import pandas as pd
import numpy as np

def _int32(x):  return int(0xFFFFFFFF & int(x))
id = 1
mode = 13
level = 1000
filename = 'output.csv'
```

```python
data = []
with open(filename, 'r') as f:
    for line in f:
        row = line.strip().split(',')
        data.append([_int32(int(s)) for s in row])
# data
```

```python
seeds = [(f'{_int32(data[0][i]-id-mode-0x65*level):X}') for i in range(1,len(data[0]))]
df = pd.DataFrame(seeds)  

csv_file = "output_real.csv"
df.to_csv(csv_file, index=False)
```
