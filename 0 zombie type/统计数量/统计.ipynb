```python
# prob
import numpy as np
import pandas as pd
from math import comb

offset = [-2,-3,0,-4]
name = ['DE','NE','PE','RE']

ans_dic = {}
for sc in range(4):
    total = 19+2-1+offset[sc]
    ans = []
    for i in range(10):
        # ans.append(f'{comb(total-i,9-i)*1.0/comb(total,9):.10f}')
        ans.append(comb(total-i,9-i)*1.0/comb(total,9))
    ans_dic[name[sc]] = ans

df = pd.DataFrame(ans_dic)
df
```
