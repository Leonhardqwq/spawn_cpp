```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns

filename = 'side_output.csv'
df = pd.read_csv(filename, header=None,index_col=0).T

plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号
```

```python
len_total = len(df)
len_sp = len(df[df['总数']>0])

num_mean_total = df['总数'].mean()
num_mean_sp = df[df['总数']>0]['总数'].mean()

weight_mean_total = df['权重'].mean()
weight_mean_sp = df[df['总数']>0]['权重'].mean()
```

```python
print(f'总数:{len_total}')
print(f'出怪关数量:{len_sp}')

print(f'所有关平均数量:{num_mean_total}')
print(f'出怪关平均数量:{num_mean_sp}')

print(f'所有关平均权重:{weight_mean_total}')
print(f'出怪关平均权重:{weight_mean_sp}')

plt.figure(figsize=(16,5))
value_counts = df[df['总数']>0]['总数'].value_counts().sort_index()  
value_counts.plot(kind='bar')
plt.title('总数')
plt.show()

value_counts = df[df['总数']>0]['权重'].value_counts().sort_index()  
value_counts.plot(kind='bar')
plt.title('权重')
plt.show()
```
